# Carregando a base de dados utilizada no Bootcamp

```
url='https://raw.githubusercontent.com/alura-cursos/agendamento-hospitalar/main/dados/A160324189_28_143_208.csv'
```

```
import pandas as pd
```

```
dados = pd.read_csv(url, encoding="ISO-8859-1", 
                    skiprows = 3, sep=";", skipfooter=12,
                    thousands=".", decimal=",")
dados.head()
```

Conseguimos plotar muito bem os dados da coluna de julho de 2020 que possui apenas valores numéricos. 

```
dados['2020/Jul'].plot(kind='bar')
```

Mas para gerar o gráfico para agosto de 2020 precisamos substituir a string ('-') por um valor numérico. Neste exemplo estamos substituindo por zero. 

```
dados.replace('-',0,inplace=True)
dados.head()
```

Além disso, precisamos converter todos os outros valores da coluna para valores numéricos. Como havia uma string na coluna o pandas carregou todos os elementos da coluna como strings.

```
dados['2020/Ago'] = pd.to_numeric(dados['2020/Ago'])
```

```
dados['2020/Ago'].plot(kind='bar')
```

Outra substituição possível é a por valoes nulos. 

```
dados.replace(0,'NaN',inplace=True)
dados.head()
```

```
dados.isnull().head()
```

```
dados.replace('NaN',float('nan'),inplace=True)
dados.head()
```

```
dados.isnull().head()
```

Mas agora vamos pegar apenas as colunas com valores não nulos.

```
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0,'Unidade da Federação')
colunas_usaveis
```

```
usaveis = dados[colunas_usaveis]
```

```
usaveis = usaveis.set_index('Unidade da Federação')
```

Remover a coluna 'Total'

```
usaveis = usaveis.drop('Total',axis=1)
```

plotar os dados do Acre.

```
usaveis.loc['12 Acre'].T.plot(figsize=(10,6))
```

Adicionando um valor nulo podemos observar como fica o gráfico neste caso.

```
usaveis['2007/Dez'] = float('nan')
```

```
usaveis.loc['12 Acre'].T.plot(figsize=(10,6))
```

```
acre = usaveis.loc['12 Acre'].T
```

```
acre['2007/Nov']
```

```
acre['2008/Jan']
```

```
acre['2007/Dez'] = 1000000.0
```

```
acre.plot(figsize=(10,6))
```

```
acre['2007/Dez'] = float('nan')
acre['2010/Dez'] = float('nan')
acre['2015/Dez'] = float('nan')
```

```
acre.plot(figsize=(10,6))
```

Resetando o índice para trabalhar com dados sequênciais.

```
acre = acre.reset_index()
```

```
acre.head()
```

```
acre = acre.drop('index',axis=1)
```

Testando o fillna como forma de substituição dos valores nulos.

```
acre.fillna(method='ffill').head()
```

Mostrando a utilização de diferentes métodos de interpolação.

```
acre.interpolate(method='linear').head(6)
```

```
acre.interpolate(method='linear').plot(figsize=(10,6))
```

```
acre.interpolate(method='spline', order=3).plot(figsize=(10,6))
```
